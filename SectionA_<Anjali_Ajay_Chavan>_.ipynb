### Q1. `is` vs `==`

- `==` operator checks if two variables have the **same value**, while `is` operator checks if they point to the **same object in memory**.

- Two variables can be `==` (same value) but `is` returns False because Python created two separate list objects.

**Example:**
```python
a = [1, 2, 3]
b = [1, 2, 3]
print(a == b)  # True  — values are equal
print(a is b)  # False — different objects in memory
```

### Q7. `append` vs `extend` vs `insert`

```python
a = [1, 2, 3]
x = [4, 5]
```

- `append(x)` adds the entire list as a single element at the end.

```python
a.append(x)
# [1, 2, 3, [4, 5]]
```
- `extend(x)` adds each element of x separately to the list.

```python
a.extend(x)
# [1, 2, 3, 4, 5]
```
- insert(i, x) inserts the entire list as a single element at a specified position.

```python

# [1, [4, 5], 2, 3]
```
In simple words:

- append() → adds one item at the end.
- extend() → adds multiple items from another iterable.
- insert() → adds one item at a specific position.

###Q10. Why is NumPy Faster than Python Lists?

NumPy is faster because it is designed for efficient numerical computations.

1. Efficient Memory Storage:
NumPy arrays store data in a continuous block of memory, which allows faster access to elements. Python lists store references to objects, making them less efficient.
2. Same Data Type:
All elements in a NumPy array have the same data type. This eliminates the need for Python to check the type of each element during calculations.
3. Vectorized Operations:
NumPy performs operations using optimized C code behind the scenes. Instead of processing elements one by one in a Python loop, it operates on the entire array at once, making computations much faster.

Example:
```python
import numpy as np

arr = np.array([1, 2, 3, 4])
arr = arr * 2
```
- This multiplies all elements at once, whereas a Python list would typically require a loop. Therefore, NumPy is faster because of efficient memory usage, fixed data types, and optimized vectorized operations.

######Q31. Find all Armstrong numbers between 1 and 9999. (An n-digit number is Armstrong if the sum of its digits each raised to the power n equals the number itself. Example: 153 = 1³ + 5³ + 3³.) Print them in a clean comma-separated list with a count at the end.

````python
count = 0

for num in range(1, 10000):

    temp = num
    digits = 0

    while temp > 0:
        digits += 1
        temp = temp // 10

    temp = num
    total = 0

    while temp > 0:
        digit = temp % 10
        total = total + digit ** digits
        temp = temp // 10

    if total == num:
        print(num, end=", ")
        count += 1

print("\nTotal Armstrong Numbers =", count)
````

######Q37. Take 10 student marks (out of 100) as input, space-separated on one line, parse with split() and map(). Without using sort() or sorted(), find: (a) maximum, (b) minimum, (c) average, (d) count of students above average, (e) the SECOND highest mark.

````python
marks = list(map(int, input("Enter 10 marks: ").split()))

maximum = marks[0]
minimum = marks[0]
sum = 0

for i in marks:
    if i > maximum:
        maximum = i

    if i < minimum:
        minimum = i

    sum = sum + i

average = sum / len(marks)

count = 0
for i in marks:
    if i > average:
        count = count + 1

highest = marks[0]
second = marks[0]

for i in marks:
    if i > highest:
        second = highest
        highest = i
    elif i > second and i != highest:
        second = i

print("Maximum =", maximum)
print("Minimum =", minimum)
print("Average =", round(average, 2))
print("Students Above Average =", count)
print("Second Highest =", second)
````

######Q38. Word Frequency Analyzer for Swiggy reviews. Take a review sentence. Find the frequency of each word (case-insensitive, "Food" and "food" should be counted together). Print words sorted by frequency in descending order. Also print the MOST frequent word and its count. Use a dict.

````python
review = input("Enter review: ")

review = review.lower()
words = review.split()

freq = {}

for word in words:
    if word in freq:
        freq[word] = freq[word] + 1
    else:
        freq[word] = 1

print("\nWord Frequencies:")

while len(freq) > 0:

    max_word = ""
    max_count = 0

    for word in freq:
        if freq[word] > max_count:
            max_count = freq[word]
            max_word = word

    print(max_word, ":", max_count)
    del freq[max_word]
````

######Q45. Caesar Cipher Encryptor. Take a message and a shift value N as input. Encrypt by shifting each letter forward by N positions in the alphabet (wrapping around, Z + 1 = A). Preserve case. Keep digits, spaces, and special characters UNCHANGED. Example: "Hello, India!" with shift 3 → "Khoor, Lqgld!"

````python
text = input("Enter message: ")
shift = int(input("Enter shift value: "))

result = ""

for ch in text:

    if ch.isupper():
        new = ord(ch) + shift

        if new > ord('Z'):
            new = new - 26

        result = result + chr(new)

    elif ch.islower():
        new = ord(ch) + shift

        if new > ord('z'):
            new = new - 26

        result = result + chr(new)

    else:
        result = result + ch

print("Encrypted Message:", result)
````

######Q46. Write a function `calculate_gst(amount, gst_rate=18, gst_inclusive=False)`:

````python
def calculate_gst(amount, gst_rate=18, gst_inclusive=False):

    if gst_inclusive == False:
        base = amount
        gst = base * gst_rate / 100
        total = base + gst

    else:
        total = amount
        base = total * 100 / (100 + gst_rate)
        gst = total - base

    return round(base, 2), round(gst, 2), round(total, 2)


print(calculate_gst(1000))
print(calculate_gst(1180, gst_inclusive=True))
print(calculate_gst(500, gst_rate=5))
````

######Q47. Write a RECURSIVE function `power(base, exp)` that computes base raised to exp, WITHOUT using the `**` operator and WITHOUT using `pow()`. Handle exp = 0 (return 1) and negative exp (return 1/power(base, -exp)). Test with: power(2, 10), power(5, 0), power(3, -2).

````python
def power(base, exp):

    if exp == 0:
        return 1

    elif exp < 0:
        return 1 / power(base, -exp)

    else:
        return base * power(base, exp - 1)


print(power(2, 10))
print(power(5, 0))
print(power(3, -2))
````
